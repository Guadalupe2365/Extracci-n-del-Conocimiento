```python
import pandas as pd
```

```python
import random
```

```python
# 1. Definir la cantidad de filas deseadas (1 millón)
cantidad = 1000000
```

```python
#2. Conjunto de base de datos reales/coherentes para la simulación en cancún
influencers_pool = ["@caribe_mexicano", "@rivera_maya_life", "@yucatan_explorer", "@tulum_diares", "@nomada_mx"]
tipos_post_pool = ["foto", "video", "reel", "carrusel"]
fechas_pool = ["2026-09-07", "2026-10-27", "2026-12-03", "2026-12-12", "2026-06-27"]
```

```python
#3. Creación del dataframe de 1,000,000 de registros usando comprensión de Listas
df_instagram = pd.DataFrame({
    "influencer": [random.choice(influencers_pool) for _ in range(cantidad)],
    "likes": [random.randint(100, 50000) for _ in range(cantidad)],
    "comentarios": [random.randint(5, 2500) for _ in range(cantidad)],
    "fecha": [random.choice(fechas_pool) for _ in range(cantidad)],
    "tipo_post": [random.choice(tipos_post_pool) for _ in range(cantidad)]
})
```
